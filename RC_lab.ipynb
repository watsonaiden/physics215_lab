```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
times = np.array([i*10 for i in range(13)])
times
```

```python
# charging
charging_voltages = np.array([.001, 3.616,6.055,7.66,8.74,9.45,9.93,10.28,10.52,10.70,10.84,10.95,11.03])
plt.plot(times, charging_voltages)
plt.title('capacitor charging curve')
plt.xlabel('time(s)')
plt.ylabel('capacitor voltage(V)')
plt.savefig('charging_capacitor.png')
```

```python
discharging_voltages = np.array([12.03, 8.15, 5.6, 3.960, 2.676, 1.881, 1.312, .919, .653, .461, .328,.234,.169])
plt.plot(times, discharging_voltages)
plt.title('capacitor discharging curve')
plt.xlabel('time(s)')
plt.ylabel('capacitor voltage(V)')
plt.savefig('discharging_capacitor.png')
```
